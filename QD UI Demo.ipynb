```python
from live_plotter import LivePlotter
from bokeh.io import output_notebook
output_notebook()

import os 
os.environ["BOKEH_ALLOW_WS_ORIGIN"] = "localhost:8889"
        
```

```python
# instrument interfacing happens here 

# define variables 
measurement_variables = [ 1, 1, -1 , -1]

# define setters
def set_0(v):
    measurement_variables[0] = v

def set_1(v):
    measurement_variables[1] = v
    
def set_2(v):
    measurement_variables[2] = v
    
def set_3(v):
    measurement_variables[3] = v

# define getters
def get_0():
    return measurement_variables[0]

def get_1():
    return measurement_variables[1]

def get_2():
    return measurement_variables[2]

def get_3():
    return measurement_variables[3]



# define measurement function 

import numpy as np 

x , y  = np.linspace(-1, 1, 80) , np.linspace(-1, 1, 80)
xv, yv = np.meshgrid(x, y, sparse=True)

from time import time
start_time = time()


def evaluate_polynomial():
         
        roots = [
        measurement_variables[0] + measurement_variables[1] * 1j,
        measurement_variables[2] + measurement_variables[3] * 1j,
            ]
        
        P =  np.product([z - xv - yv*1j for z in roots], axis = 0)
        
        
        return np.abs(P + P * 0.01 * np.sqrt(time() - start_time)* np.random.rand(* P.shape) ) 
        
```

```python
plot = LivePlotter("Plot name")

plot.set_measurement_function( evaluate_polynomial )

plot.set_callback_time(500)

plot.add_variable_widget(
                        set_0 , 
                        get_0 ,  
                        variable_name="Real part of first root", 
                        unit="", 
                        initial_value=1,
                        initial_stepsize = None, # None defaults to 10*resolution
                        variable_resolution  = 2e-4,
                        stepsize_lower_bound = 1e-6,
                        stepsize_upper_bound = 1,
                        )
plot.add_variable_widget(
                        set_1 , 
                        get_1 ,  
                        variable_name="Imaginary part of first root", 
                        unit="", 
                        initial_value=1,
                        initial_stepsize = None,  
                        variable_resolution  = 2e-4,
                        stepsize_lower_bound = 1e-6,
                        stepsize_upper_bound = 1,
                        )
plot.add_variable_widget(
                        set_2 , 
                        get_2 ,  
                        variable_name="Real part of second root", 
                        unit="", 
                        initial_value=-1,
                        initial_stepsize = None,  
                        variable_resolution  = 2e-4,
                        stepsize_lower_bound = 1e-6,
                        stepsize_upper_bound = 1,
                        )
plot.add_variable_widget(
                        set_3 , 
                        get_3 ,  
                        variable_name="Imaginary part of second root", 
                        unit="", 
                        initial_value=-1,
                        initial_stepsize = None,  
                        variable_resolution  = 2e-4,
                        stepsize_lower_bound = 1e-6,
                        stepsize_upper_bound = 1,
                        )
 
plot.set_plot_dimensions(-1 , -1, 1, 1)

plot.set_x_label("X label")

plot.set_y_label("Y label")
```

```python
plot.show()
```
